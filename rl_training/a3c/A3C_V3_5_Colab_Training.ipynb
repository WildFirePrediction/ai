## 4️⃣ Training Configuration

````python
# Training hyperparameters
NUM_WORKERS = 4           # 4-8 for Colab (parallel training)
MAX_EPISODES = 1000       # Total episodes to train
TEMPORAL_WINDOW = 5       # Full 5 timesteps (architecture plan)
MAX_GRID_CELLS = 150000   # 387×387 grids
MIN_EPISODE_LENGTH = 4    # Quality filter
LEARNING_RATE = 7e-5
ENTROPY_COEF = 0.015

# WandB settings
WANDB_PROJECT = "wildfire-prediction-colab"
WANDB_RUN_NAME = f"v3.5-colab-w{NUM_WORKERS}-tw{TEMPORAL_WINDOW}"

print(f"Configuration:")
print(f"  Workers: {NUM_WORKERS}")
print(f"  Episodes: {MAX_EPISODES}")
print(f"  Temporal window: {TEMPORAL_WINDOW} timesteps")
print(f"  Max grid: {MAX_GRID_CELLS:,} cells")
print(f"  Expected time: 1-2 hours")
print(f"  Target IoU: 47-50%")
````

## 🆘 Troubleshooting

### OOM Error
```python
# Reduce workers
NUM_WORKERS = 2

# Reduce grid size
MAX_GRID_CELLS = 100000
```

### Slow Training
```python
# Use fewer environments
--max-envs 500

# Relax quality filter
MIN_EPISODE_LENGTH = 3
```

### No GPU Allocated
```
Runtime → Change runtime type → Hardware accelerator: GPU
```
